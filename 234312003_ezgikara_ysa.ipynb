```python
# -*- coding: utf-8 -*-
# 1. GEREKLİ KÜTÜPHANELERİ YÜKLEME

import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121, MobileNet
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import (classification_report, confusion_matrix,
                           roc_curve, auc, accuracy_score, precision_score,
                           recall_score, f1_score, roc_auc_score)
from google.colab import drive
import gc


# 2. GENEL AYARLAR VE GPU OPTİMİZASYONU

class Ayarlar:
    """Genel ayarlar için sınıf"""

    def __init__(self):
        self.IMG_BOYUTU = (224, 224)  # Ödevde verilen örnek boyut
        self.BATCH_SIZE = 32  # Optimize edilmiş batch size
        self.EPOCHS = 100    # Ödevde belirtilen epoch sayısı

    def gpu_ayarla(self):
        """GPU ayarlarını optimize eder"""
        gpus = tf.config.experimental.list_physical_devices('GPU')
        if gpus:
            try:
                for gpu in gpus:
                    tf.config.experimental.set_memory_growth(gpu, True)
                print("GPU bellek büyümesi ayarlandı")
            except RuntimeError as e:
                print(e)

    def bellek_temizle(self):
        """Bellek optimizasyonu yapar"""
        gc.collect()
        tf.keras.backend.clear_session()

    def drive_baglan(self):
        """Google Drive'a bağlanır"""
        drive.mount('/content/drive')


# 3. VERİ SETİ HAZIRLAMA

class VeriHazirlayici:
    """Veri seti hazırlama işlemleri için sınıf"""

    def __init__(self, veri_dizini, ayarlar):
        self.veri_dizini = veri_dizini
        self.ayarlar = ayarlar

    def veri_yukleyicileri_olustur(self):
        """Eğitim, validasyon ve test veri yükleyicilerini oluşturur"""
        # Test seti için ayrı generator (augmentation yok)
        test_datagen = ImageDataGenerator(
            rescale=1.0/255,
            validation_split=0.2  # İlk %20'yi test için ayır
        )

        # Test setini oluştur
        test_generator = test_datagen.flow_from_directory(
            self.veri_dizini,
            target_size=self.ayarlar.IMG_BOYUTU,
            batch_size=self.ayarlar.BATCH_SIZE,
            class_mode='categorical',
            subset='validation',  # Validation split'i test seti olarak kullan
            shuffle=False
        )

        # Eğitim ve validasyon için veri artırmalı generator
        train_datagen = ImageDataGenerator(
            rescale=1.0/255,
            rotation_range=40,
            width_shift_range=0.1,
            height_shift_range=0.1,
            shear_range=0.1,
            zoom_range=0.1,
            horizontal_flip=True,
            fill_mode='nearest',
            validation_split=0.2  # Kalan %80'in %20'si validasyon
        )

        # Eğitim setini oluştur
        train_generator = train_datagen.flow_from_directory(
            self.veri_dizini,
            target_size=self.ayarlar.IMG_BOYUTU,
            batch_size=self.ayarlar.BATCH_SIZE,
            class_mode='categorical',
            subset='training'
        )

        # Validasyon setini oluştur
        validation_generator = train_datagen.flow_from_directory(
            self.veri_dizini,
            target_size=self.ayarlar.IMG_BOYUTU,
            batch_size=self.ayarlar.BATCH_SIZE,
            class_mode='categorical',
            subset='validation'
        )

        return train_generator, validation_generator, test_generator


# 4. MODEL OLUŞTURMA

class ModelMimarisi:
    """Model mimarisi oluşturma işlemleri için sınıf"""

    def __init__(self, ayarlar):
        self.ayarlar = ayarlar

    def model_olustur(self, model_adi="DenseNet121"):
        """Transfer learning modeli oluşturur"""
        base_model = self._temel_model_getir(model_adi)
        base_model.trainable = False

        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        x = Dropout(0.5)(x)
        # Ödevde belirtilen dense katmanları
        x = Dense(256, activation='relu')(x)
        x = Dense(256, activation='relu')(x)
        x = Dense(128, activation='relu')(x)
        x = Dense(64, activation='relu')(x)
        predictions = Dense(2, activation='softmax')(x)

        model = Model(inputs=base_model.input, outputs=predictions)

        # ADAM optimizer, ödevde belirtilen lr=0.001
        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def _temel_model_getir(self, model_adi):
        """Seçilen temel modeli yükler"""
        if model_adi == "DenseNet121":
            return DenseNet121(
                weights='imagenet',
                include_top=False,
                input_shape=(*self.ayarlar.IMG_BOYUTU, 3)
            )
        else:  # MobileNet (V1)
            return MobileNet(
                weights='imagenet',
                include_top=False,
                input_shape=(*self.ayarlar.IMG_BOYUTU, 3)
            )


# 5. MODEL EĞİTİMİ

class ModelEgitici:
    """Model eğitim işlemleri için sınıf"""

    def __init__(self, ayarlar):
        self.ayarlar = ayarlar
        self.callbacks = self._callbacks_olustur()

    def _callbacks_olustur(self):
        """Callback fonksiyonlarını oluşturur"""
        return [
            # Early stopping
            EarlyStopping(
                monitor='val_loss',
                patience=25,
                restore_best_weights=True
            ),
            # ReduceLROnPlateau
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.1,
                patience=5
            ),
            # En iyi modeli kaydet
            ModelCheckpoint(
                filepath='best_model.weights.h5',  # Değiştirilen uzantı
                monitor='val_loss',
                save_best_only=True,
                save_weights_only=True
            )
        ]

    def egit(self, model, train_generator, validation_generator):
        """Modeli eğitir"""
        history = model.fit(
            train_generator,
            steps_per_epoch=len(train_generator),
            validation_data=validation_generator,
            validation_steps=len(validation_generator),
            epochs=self.ayarlar.EPOCHS,
            callbacks=self.callbacks
        )
        return history


# 6. PERFORMANS ANALİZİ

class PerformansAnalizci:
    """Model performans analizi işlemleri için sınıf"""

    def metrikleri_hesapla(self, y_true, y_pred, y_pred_probs):
        """Tüm performans metriklerini hesaplar"""
        conf_matrix = confusion_matrix(y_true, y_pred)
        tn, fp, fn, tp = conf_matrix.ravel()

        # Ödevde istenen tüm metrikler
        specificity = tn / (tn + fp)

        metrikler = {
            'Doğruluk (Accuracy)': accuracy_score(y_true, y_pred),
            'Hassasiyet (Precision)': precision_score(y_true, y_pred),
            'Duyarlılık (Recall)': recall_score(y_true, y_pred),
            'Özgüllük (Specificity)': specificity,
            'F1-score': f1_score(y_true, y_pred),
            'AUC-ROC': roc_auc_score(y_true, y_pred_probs)
        }

        return metrikler, conf_matrix

    def sonuclari_goster(self, history, conf_matrix, metrikler, model_adi, y_true, y_pred_probs):
        """Tüm sonuçları görselleştirir"""
        # 1. Eğitim/Validasyon Grafikleri
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Eğitim')
        plt.plot(history.history['val_accuracy'], label='Validasyon')
        plt.title(f'{model_adi} - Doğruluk')
        plt.ylabel('Doğruluk')
        plt.xlabel('Epoch')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Eğitim')
        plt.plot(history.history['val_loss'], label='Validasyon')
        plt.title(f'{model_adi} - Kayıp')
        plt.ylabel('Kayıp')
        plt.xlabel('Epoch')
        plt.legend()

        plt.tight_layout()
        plt.show()

        # 2. Karmaşıklık Matrisi
        plt.figure(figsize=(6, 6))
        sns.heatmap(
            conf_matrix,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=['Maskesiz', 'Maskeli'],
            yticklabels=['Maskesiz', 'Maskeli']
        )
        plt.title(f'Karmaşıklık Matrisi - {model_adi}')
        plt.show()

        # 3. ROC Eğrisi
        fpr, tpr, _ = roc_curve(y_true, y_pred_probs)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(6, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                label=f'ROC Eğrisi (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('Yanlış Pozitif Oranı')
        plt.ylabel('Doğru Pozitif Oranı')
        plt.title(f'ROC Eğrisi - {model_adi}')
        plt.legend(loc="lower right")
        plt.show()

        # 4. Metrikleri yazdır
        print(f"\n{model_adi} Metrikleri:")
        for metrik, deger in metrikler.items():
            print(f"{metrik}: {deger:.4f}")

    def hiperparametre_analizi_goster(self):
        """Hiperparametre analizini gösterir"""
        print("\nHiperparametre Analizi ve Model Seçimi:")
        print("----------------------------------------")
        print("1. Model Seçimi:")
        print("   - DenseNet121: Daha derin ve karmaşık yapısıyla yüksek doğruluk")
        print("   - MobileNet: Daha hafif yapısıyla hızlı eğitim ve tahmin")
        print("\n2. Hiperparametreler:")
        print("   - Batch Size: 32 (GPU bellek optimizasyonu için)")
        print("   - Learning Rate: 0.001 (Adam optimizer için ideal başlangıç)")
        print("   - Early Stopping Patience: 25 (Yeterli öğrenme süresi için)")
        print("   - ReduceLROnPlateau:")
        print("     * Factor: 0.1 (Kademeli öğrenme hızı düşüşü)")
        print("     * Patience: 5 (Hızlı adaptasyon için)")
        print("\n3. Model Katmanları:")
        print("   - Dropout: 0.5 (Overfitting kontrolü)")
        print("   - Dense Katmanları: 256->256->128->64 (Kademeli boyut azaltma)")


# 7. ANA PROGRAM

def main():
    """Ana program akışı"""
    # Ayarlar ve optimizasyonlar
    ayarlar = Ayarlar()
    ayarlar.gpu_ayarla()
    ayarlar.bellek_temizle()
    ayarlar.drive_baglan()

    # Veri seti hazırlama
    veri_dizini = '/content/drive/MyDrive/data'  # Kendi yolunuzu belirtin
    veri_hazirlayici = VeriHazirlayici(veri_dizini, ayarlar)
    train_generator, validation_generator, test_generator = veri_hazirlayici.veri_yukleyicileri_olustur()

    # Model oluşturma ve eğitim nesneleri
    model_mimarisi = ModelMimarisi(ayarlar)
    model_egitici = ModelEgitici(ayarlar)
    performans_analizci = PerformansAnalizci()

    # Her iki modeli de eğit ve değerlendir
    modeller = ["DenseNet121", "MobileNet"]  # MobileNet V1 kullanıldı
    sonuclar = {}

    for model_adi in modeller:
        print(f"\n{model_adi} modeli eğitiliyor...")

        # Model oluştur ve eğit
        model = model_mimarisi.model_olustur(model_adi)
        history = model_egitici.egit(model, train_generator, validation_generator)

        # Test seti üzerinde tahminler
        print(f"\n{model_adi} test seti üzerinde değerlendiriliyor...")
        y_pred_probs = model.predict(test_generator)
        y_pred = np.argmax(y_pred_probs, axis=1)
        y_true = test_generator.classes

        # Performans metrikleri
        metrikler, conf_matrix = performans_analizci.metrikleri_hesapla(
            y_true, y_pred, y_pred_probs[:, 1]
        )

        # Sonuçları göster
        performans_analizci.sonuclari_goster(
            history, conf_matrix, metrikler, model_adi, y_true, y_pred_probs[:, 1]
        )

        # Sonuçları sakla
        sonuclar[model_adi] = {
            'metrikler': metrikler,
            'conf_matrix': conf_matrix,
            'history': history.history
        }

        # Belleği temizle
        ayarlar.bellek_temizle()

    # Hiperparametre analizi ve model karşılaştırması
    print("\nModel Karşılaştırması:")
    print("----------------------")
    metrik_adlari = list(sonuclar[modeller[0]]['metrikler'].keys())

    for metrik in metrik_adlari:
        print(f"\n{metrik}:")
        for model in modeller:
            deger = sonuclar[model]['metrikler'][metrik]
            print(f"{model}: {deger:.4f}")

    # Hiperparametre analizi
    performans_analizci.hiperparametre_analizi_goster()

    print("\nEğitim tamamlandı! Sonuçlar başarıyla kaydedildi.")

if __name__ == "__main__":
    main()
```
